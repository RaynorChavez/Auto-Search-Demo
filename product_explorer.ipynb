```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")

# Load the CSV data into a DataFrame
#file_path = "./data/bisoncoolers_products.csv"  
file_path = "./data/taylorstitch_products.csv"  
#file_path = "./data/decathlon_products.csv"  
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
pd.set_option('max_colwidth', None)  # Display the full content of the column without truncation
df.head(1)
```

```python
row_number = 10  # for the 3rd row
print(df.iloc[row_number])
```

```python
# Check the number of unique products
unique_products = df['Name'].nunique()
print(f"Number of unique products: {unique_products}")

# Check the number of unique categories
unique_categories = df['Category'].nunique()
print(f"Number of unique categories: {unique_categories}")

# Average price of products
average_price = df['Price'].mean()
print(f"Average product price: ${average_price:.2f}")
```

```python
# Price distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['Price'], bins=5, color='skyblue')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Number of Products')
plt.show()
```

```python
# Product distribution across categories
plt.figure(figsize=(14, 7))
sns.countplot(data=df, y='Category', order=df['Category'].value_counts().index, palette="viridis")
plt.title('Number of Products per Category')
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.show()
```
